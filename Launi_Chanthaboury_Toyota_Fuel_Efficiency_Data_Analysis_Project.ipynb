## Analyses and Visualizations
In this visualization, I decided to look at the three most popular car manufacturers within the "Big Three." The Big Three refers to the countries that producing the most cars, Japan, Germany, and the USA. I looked at the different brands highway MPG fuel efficiency in order to see how well they performed against each other. Although Volkswagen performed significantly better than Toyota, I wanted to continue to dig deeper with Toyota and figure out what made them the most popular car manufacturer overall.


```python
# Top 3 Brand's highway mpg
ford_mask = df.loc[:, 'Identification.Make'].str.contains('Ford')
toyota_mask = df.loc[:, 'Identification.Make'].str.contains('Toyota')
volks_mask = df.loc[:, 'Identification.Make'].str.contains('Volkswagen')

top_3 = df[ford_mask+ toyota_mask + volks_mask]
top_3_hue = top_3.loc[:, 'Identification.Make']

sns.lineplot(x = 'Identification.Year', y = 'Fuel Information.Highway mpg', data = df, ci = None, hue = top_3_hue)
plt.xticks([2010, 2011, 2012])
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.title('Big 3 Highway MPG overtime')


plt.show()
```

In the visualization below, I looked at the highway MPG for all Toyota vehicles. With this lineplot, I used manual/auto transmission as the hue, so the viewer could see out of all Toyota vehicles, which type of transmission worked the best. This was done so I could see general view as to what transmission performs better. Through the graph, we can see that it is manual that significantly outperforms automatic.

It is important to see the spike in 2011, for both auto and manual transmissions. This is a common trend that I have noticed throughout my research.


```python
 # Toyota's highway mpg

toyota_mask = df.loc[:, 'Identification.Make'] == 'Toyota'
toyota_only = df[toyota_mask]

sns.lineplot(x = 'Identification.Year', y = 'Fuel Information.Highway mpg', data = toyota_only, ci = None, hue = 'Identification.Classification')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xticks([2010, 2011, 2012])
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.title('Toyota Highway MPG overtime')
plt.show()
```

In the next few visualizations, I made a mask for each different vehicle type. I looked at Toyota's sedans, SUVs, Minivan, compact cars, and trucks. Below is a lineplot of sedans and its highway MPG over the years. Although the data for manual transmission is limited, we can still see it outperforming automatic again.


```python
# Toyota Sedans

# Sedan Masking
camry_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('Camry')
corolla_mask = toyota_only.loc[:,'Identification.ID'].str.contains('Corolla')
avalon_mask = toyota_only.loc[:,'Identification.ID'].str.contains('Avalon')


sedan_df = toyota_only[camry_mask + corolla_mask + avalon_mask]


sns.lineplot(x = 'Identification.Year', y = 'Fuel Information.Highway mpg', data = sedan_df, ci = None, hue = 'Identification.Classification')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xticks([2010, 2011, 2012])
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.title('Toyota Sedan Highway MPG overtime')

plt.show()
```

Below is a graph of Toyota's SUVs and its highway MPG. It is interesting to see that SUVs follow the exact opposite trend when it comes to manual and automatic transmissions. I am not completely sure as to why, but I think it may be because of the significant weight increase when it comes to SUVs compared to sedans.

```python
# SUV Masking
four_runner_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('4Runner')
fj_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('FJ')
highlander_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('Highlander')
landcruiser_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('Land Cruiser')
rav4_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('RAV4')
sequoia_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('Sequoia')
venza_mask = toyota_only.loc[:, 'Identification.ID'].str.contains('Venza')

suv_df = toyota_only[four_runner_mask + fj_mask + highlander_mask + landcruiser_mask + rav4_mask + sequoia_mask + venza_mask]

sns.lineplot(x = 'Identification.Year', y = 'Fuel Information.Highway mpg', data = suv_df, ci = None, hue = 'Identification.Classification')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xticks([2010, 2011, 2012])
plt.xlabel('Year')
plt.ylabel('Highway MPG')
plt.title('Toyota SUV Highway MPG overtime')

plt.show()
```

This is a visualization of compact cars from Toyota. These are by far the must fuel efficient types of vehicles when it comes to looking at Toyota cars. They outperform each vehicle type, and the most fuel efficient car would be a compact car that is also has a manual transmission.
